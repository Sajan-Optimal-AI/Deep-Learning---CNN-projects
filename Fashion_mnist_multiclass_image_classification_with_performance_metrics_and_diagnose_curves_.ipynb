```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
Train_Data = pd.read_csv('/content/fashion-mnist_train.csv')
Test_Data = pd.read_csv('/content/fashion-mnist_test.csv')
print(Train_Data.shape,Test_Data.shape)
```

```python
Train_Data.head()
```

```python
Test_Data.head()
```

```python
# Convert dataframe (csv) into numpy float array 
train_numpy = np.array(Train_Data,dtype='float32')
test_numpy = np.array(Test_Data,dtype='float32')
# Split features and labels from both test and train numpy data
X_train = train_numpy[:,1:]
y_train = train_numpy[:,0]
X_test = test_numpy[:,1:]
y_test = test_numpy[:,0]
```

```python
# Normalize the split..ie fashion_mnist images are greyscale ie.pixels ranges between(0 to 255).So In order to normalize divide numby array data by 255.
X_train_N = X_train/255
X_test_N = X_test/255
# No need to normalize test data........coz its not pixels and its class labels
```

```python
# Now split the train and normalized data into once again training and validation data.
x_train , x_validate , y_train , y_validate = train_test_split(X_train_N,y_train,test_size=0.2,random_state=12345)   # 12345 ???????
```

```python
# Pick the image for reference
image = X_train[100,:].reshape(28,28)
plt.imshow(image)
plt.show
```

```python
# Reshape features 
x_train = x_train.reshape(x_train.shape[0],28,28,1)
x_test = X_test_N.reshape(X_test_N.shape[0],28,28,1)
x_validate = x_validate.reshape(x_validate.shape[0],28,28,1)
print('x_train shape: {}'.format(x_train.shape))
print('x_test shape: {}'.format(x_test.shape))
print('x_validate shape: {}'.format(x_validate.shape))
```

```python
# Build CNN Model
import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten , Dropout
from keras.optimizers import Adam
Model = Sequential()
Model.add(Conv2D(filters=32,kernel_size=(3,3),activation='relu',input_shape=(28,28,1)))
Model.add(MaxPooling2D(pool_size=(2,2)))
Model.add(Dropout(0.2))
Model.add(Flatten())
Model.add(Dense(units=32, activation='relu'))
Model.add(Dense(units=10 , activation='softmax'))
Model.compile(optimizer=Adam(lr=0.001), loss='sparse_categorical_crossentropy',metrics=['accuracy'])
print(Model.summary())
```

```python
# Evaluate the model
Results = Model.fit(x_train,y_train,batch_size=512,epochs=30,verbose=1,validation_data=(x_validate,y_validate))
```

```python
Score = Model.evaluate(x_test,y_test,verbose=1) # Avearge Accuracy and loss
```

```python
from sklearn.metrics import accuracy_score , classification_report,confusion_matrix
```

```python
# Diagnose curves # for accuracy
plt.plot(Results.history['accuracy'])
plt.plot(Results.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend(['Train_set','Validation_set'], loc='lower right')
plt.show
```

```python
# Diagnose curves # for loss
plt.plot(Results.history['loss'])
plt.plot(Results.history['val_loss'])
plt.title('Model loss')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend(['Train_set','Validation_set'], loc='upper right')
plt.show
```

```python
# Evaluate Performance metrics
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix 
from sklearn import preprocessing
predictions = Model.predict(x_test)
predictions = np.argmax(predictions, axis=-1)
le = preprocessing.LabelEncoder()
le.fit(predictions)
predictions = le.inverse_transform(predictions)
cm = confusion_matrix(y_test,predictions)
cr = classification_report(y_test,predictions)
print(cm)
print(cr)
```
